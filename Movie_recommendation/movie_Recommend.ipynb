```python
df_movies = pd.read_csv('Data/tmdb_5000_movies.csv')
df_credits = pd.read_csv('Data/tmdb_5000_credits.csv')
print(f'the columns of movies : {df_movies.columns}', end=" ")
print(f'the columns of credits : {df_credits.columns}', end= " ")
```

```python
movies = df_movies.merge(df_credits,on='title')
movies.shape
```

```python
## genres, id, keywords,popularity,release_Date,revenue, cast, overview, crew
movies.columns
```

```python
## selected columns 
movie_slc = movies[['genres','id','title','keywords','popularity','overview','release_date','cast','crew']]
movie_slc.head()
```

```python
## duplciated value check 
movie_slc.duplicated().sum()
```

```python
movie_cpy = movie_slc.copy()
movie_cpy.dropna(inplace=True)
```

```python
movie_cpy.reset_index(inplace=True)
```

```python
movie_cpy.tail()
```

```python
import ast
def convert_obj(obj_lst):
    l = []
    for i in ast.literal_eval(obj_lst):
        l.append(i["name"])
    return l

 
```

```python

movie_cpy['movie_genre_names']=  movie_cpy['genres'].apply(convert_obj)
movie_cpy['keywords_name'] = movie_cpy['keywords'].apply(convert_obj)
len(ast.literal_eval(movie_cpy['cast'][3]))
```

```python
# to get the cast names we need to get 3 cast names from the movie community
def convert_cast(obj):
    l=[]
    counter=1
    for j in ast.literal_eval(obj):
        if counter!=4:
            l.append((counter,j['name']))
            counter=counter+1
        else:
            break
        

    return l

movie_cpy['cast_name'] = movie_cpy['cast'].apply(convert_cast)
movie_cpy['cast_name'][0]
```

```python
def convert_crew(obj):
    m=[]
    for l in ast.literal_eval(obj):
        if l['job']=='Director':
            m.append(l['name'])
            break
    return m

movie_cpy['director_name'] = movie_cpy['crew'].apply(convert_crew)
movie_cpy.head()
```

```python
#movie_cpy['overview'] = movie_cpy['overview'].apply(lambda x:x.split())
[(cn,"".join(nme.split()))   for cn, nme in movie_cpy['cast_name'][0]]
```

```python
movie_cpy['cast_name']=movie_cpy['cast_name'].apply(lambda x : [(cn,"".join(nme.split()))  for cn,nme in x])
movie_cpy['director_name'] =  movie_cpy['director_name'].apply(lambda x: ["".join(j.split()) for j in x])
movie_cpy.head()
```

```python
for j in  movie_cpy['director_name']:
    for m in j:
        print("".join(m.split()))
```

```python
movie_cpy["movie_genre_names"] = movie_cpy['movie_genre_names'].apply(lambda x: ["".join(m.split(" ")) for m in x])
movie_cpy["keywords_name"] = movie_cpy["keywords_name"].apply(lambda x: ["".join(m.split(" ")) for m in x])
```

```python
movie_cpy.head()
```

```python
movie_cpy['overview'].head()
```

```python
movie_cpy['overview'] = movie_cpy['overview'].apply(lambda x:x.split())
movie_cpy['cast_name'] = movie_cpy['cast_name'].apply(lambda x:[val for ky,val in x])
movie_cpy
```

```python
movie_cpy['tags'] = movie_cpy['overview']+movie_cpy['keywords_name']+movie_cpy['cast_name']+movie_cpy['director_name']
movie_cpy.head()
```

```python
new_mvie = movie_cpy[['id','title','tags']]
new_mvie['tags']= new_mvie['tags'].apply(lambda x:" ".join(x))
new_mvie.head()
```

```python
new_mvie['tags']= new_mvie['tags'].apply(lambda x:x.lower())
new_mvie
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')

```

```python
vectores = cv.fit_transform(new_mvie['tags']).toarray()
vectores
```

```python
cnter=0
fe_50=[]
for fe in list(cv.get_feature_names_out()):
    if cnter!=50:
        fe_50.append(fe)
        cnter=cnter+1
    else:
        break

print(fe_50)


```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemming_words(txt):
    tx=[]
    for i in txt.split():
        tx.append(ps.stem(i))

    return " ".join(tx)
```

```python
new_mvie['tags'] = new_mvie['tags'].apply(stemming_words)
new_mvie['tags'][0]
```

```python
fe = list(cv.get_feature_names_out())
fe_100 = fe[:100]
fe_100
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectores).shape
```

```python
## here we can see that its giving us matrix , where for its calculating similarity between one movie while
#comparing with other 4800 movies 
similarity_mtrix = cosine_similarity(vectores)
similarity_mtrix[2]
```

```python
sorted(list(enumerate(similarity_mtrix[1])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def Recommend_Movie(movie):
    movie_index= new_mvie[new_mvie['title']==movie].index[0]
    movie_lst=sorted(list(enumerate(similarity_mtrix[movie_index])),reverse=True,key=lambda x:x[1])[1:6]
    for mve in movie_lst:
        print(new_mvie.iloc[mve[0]]['title'])
```
